#  Question 1 : Gram-Schmidt Orthogonalization

#### Generate three random vectors in $R^3$ by

- `u1 = np.array([[4], [9], [6]])`
- `u2 = np.array([[6], [8], [8]])`
- `u3 = np.array([[0], [9], [6]])`

- Check whether they are linearly independent by calculating `np.linalg.matrix_rank(np.block([u1, u2, u3]))`.
- If the answer is not 3, then generate a new random set of vectors and calculate the rank. Repeat until the rank is 3. Now use these vectors in the following calculations.

```python
u1 = np.array([[4], [9], [6]])
u2 = np.array([[6], [8], [8]])
u3 = np.array([[0], [9], [6]])

rank = np.linalg.matrix_rank(np.block([u1, u2, u3]))

if rank == 3:
    print("Vectors are linearly independent")
else:
    print("vectors are linearly dependent")
    while True:
        u1 = np.random.randint(0, 10, (3, 1))
        u2 = np.random.randint(0, 10, (3, 1))
        u3 = np.random.randint(0, 10, (3, 1))

        rank = np.linalg.matrix_rank(np.block([u1, u2, u3]))

        if rank == 3:
            print("New linearly independent vectors found:")
            break

print("--------u1------------")
print(u1)
print("---------------------")
print("--------u2------------")
print(u2)
print("---------------------")
print("--------u3------------")
print(u3)
print("---------------------")
print("--------rank([u1, u2, u3])------------")
print(rank)
print("---------------------")
```

#### (b) Now use the vectors u1, u2, u3 to obtain an orthogonal basis for R3, following the Gram-Schmidt algorithm. Set `v1 = u1`. Obtain v2 by removing the component of u2 in the direction v1:


\\[v_2 = u_2 - \frac{v_1 \cdot u_2}{v_1 \cdot v_1}v_1\\]

- Calculate the dot product to check that the vectors v1 and v2 are mutually orthogonal (within a negligible numerical error). Also add v2 to your graphics window as a dashed-dotted green line.

```python
# Gram-Schmidt
v1 = u1
v2 = u2 - ((np.dot(u2.T, v1) / np.dot(v1.T, v1)) * v1)

dot_product = np.dot(v1.T, v2)

r = np.arange(0, 1, 0.05)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot3D(r * u1[0], r * u1[1], r * u1[2], 'r:', label='v1')
ax.plot3D(r * v2[0], r * v2[1], r * v2[2], 'y--', label='v2')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('Visualization of Vectors')
ax.legend()
plt.show()

print("--------v1------------")
print(v1)
print("---------------------")
print("--------v2------------")
print(v2)
print("---------------------")
print("--------v1 . v2------------")
print(dot_product)
print("---------------------")
```

#### (c) The last step in the Gram-Schmidt algorithm is to rescale the vectors v1, v2, v3 to obtain an orthonormal basis for R3:

- `w1 = v1 / np.linalg.norm(v1)`
- `w2 = v2 / np.linalg.norm(v2)`
- `w3 = v3 / np.linalg.norm(v3)`

#### Define the matrix Q = [w1, w2, w3] and give written answers to the following questions.



- What is the inverse matrix Q Now.

#### =======> The inverse for Q is Equal to Q^T

```python
w1 = v1 / np.linalg.norm(v1)
w2 = v2 / np.linalg.norm(v2)
w3 = v3 / np.linalg.norm(v3)
Q = np.block([[w1, w2, w3]])

# Q^T times Q (check orthonormality)
Q_transpose_Q = np.dot(Q.T, Q)

# The inverse inverse of matrix 
inv_Q = np.linalg.inv(Q)

print("--------w1------------")
print(w1)
print("---------------------")
print("--------w2------------")
print(w2)
print("---------------------")
print("--------w3------------")
print(w3)
print("---------------------")
print("--------Q------------")
print(Q)
print("---------------------")
print("--------Q.T @ Q------------") #Must return an Identitiy Matrix because the (Q^T=Q^(-1))
print(Q_transpose_Q)
print("---------------------")
print("--------inv(Q)------------")
print(inv_Q)
print("---------------------")
```

#### Now calculate the characteristic polynomial p(t) of your matrix A by

- Symbolic computations are done with `Symbol` class in `sympy`. Read the [documentation](https://docs.sympy.org/latest/gotchas.html#introduction) for many different examples.

#### Diagonalization Property Verification

The diagonalization of matrix \( A \). If \( A = PDP^{-1} \) where \( P \) is the matrix of eigenvectors, \( D \) is a diagonal matrix composed of eigenvalues, and \( P^{-1} \) is the inverse of matrix \( P \), then for powers of \( A \):

- \( A^5 = PD^5P^{-1} \)
- \( A^{10} = PD^{10}P^{-1} \)

```python
_A = P @ D @ np.linalg.inv(P)

A_5 = np.linalg.matrix_power(A, 5)
A_10 = np.linalg.matrix_power(A, 10)

symbo_A_5 = P @ np.linalg.matrix_power(D, 5) @ np.linalg.inv(P)
symbo_A_10 = P @ np.linalg.matrix_power(D, 10) @ np.linalg.inv(P)

print("------------A------------------")
print(A)
print("-------------------------------")
print("------------P @ D @ np.linalg.inv(P)------------------")
print(_A)
print("-------------------------------")
print("------------A^5------------------")
print(A_5)
print("-------------------------------")
print("------------P @ D^5 @ P^-1------------------")
print(symbo_A_5)
print("-------------------------------")
print("------------A^10------------------")
print(A_10)
print("-------------------------------")
print("------------P @ D^10 @ P^-1------------------")
print(symbo_A_10)
print("-------------------------------")
```
